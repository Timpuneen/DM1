```python
df = pd.DataFrame({
    'A': [10, 12, 14, 15, 16, 18, 20, 100],  # Outlier at 100
    'B': [20, 22, 24, 26, 28, 30, 32, 200]   # Outlier at 200
})
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
```

```python
df_cleaned = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
print("Orig:")
print(df)
print("\nCleaned:")
print(df_cleaned)
```

```python
#Q1 (25-й процентиль): значение, ниже которого находятся 25% данных.
#Q3 (75-й процентиль): значение, ниже которого находятся 75% данных.
#IQR (интерквартильный размах): диапазон между 25-м и 75-м процентилями, который показывает, где находится центральная часть данных.

#(Q1 - 1.5 * IQR): нижняя граница, за которой значения считаются выбросами.
#(Q3 + 1.5 * IQR): верхняя граница, за которой значения считаются выбросами.

# .any(axis=1): проверяет, есть ли хотя бы одно значение в строке, которое является выбросом.
# ~: оператор инверсии, который позволяет оставить только строки, не содержащие выбросов.
```

```python
# SimpleImputer для заполнения пропущенных значений, StandardScaler для масштабирования числовых признаков, 
# и OneHotEncoder для кодирования категориальных переменных.
```

```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
```

```python
# Пример данных
data = pd.DataFrame({
    'age': [25, np.nan, 35, 45],
    'salary': [50000, 60000, np.nan, 80000],
    'city': ['New York', 'Paris', 'Berlin', np.nan]
})

# Определяем признаки
numerical_features = ['age', 'salary']
categorical_features = ['city']
```

```python
# Шаги для числовых признаков
numerical_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),   # Заполняем пропуски средними значениями
    ('scaler', StandardScaler())                   # Масштабируем данные
])
```

```python
# Шаги для категориальных признаков
categorical_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),  # Заполняем пропуски наиболее частыми значениями
    ('onehot', OneHotEncoder(handle_unknown='ignore'))     # Кодируем категориальные переменные
])
```

```python
# Объединяем числовые и категориальные пайплайны
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_pipeline, numerical_features),
        ('cat', categorical_pipeline, categorical_features)
    ]
)
```

```python
# Создаем полный пайплайн
pipeline = Pipeline(steps=[('preprocessor', preprocessor)])

# Пример использования: трансформируем данные
processed_data = pipeline.fit_transform(data)

print(processed_data)
```

```python
# SimpleImputer: Заполняет пропущенные значения в числовых и категориальных данных.
# StandardScaler: Применяется для масштабирования числовых данных.
# OneHotEncoder: Кодирует категориальные переменные в бинарные (один из n).
# ColumnTransformer: Позволяет применять разные трансформации к числовым и категориальным данным одновременно.
```

```python
# Numerical Pipeline: Заполняет пропуски средними значениями и масштабирует числовые признаки.
# Categorical Pipeline: Заполняет пропуски наиболее частыми значениями и выполняет One-Hot-Encoding для категориальных данных.
# Pipeline: Включает в себя все шаги предобработки и применяется ко всему набору данных.
```

```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# Пример данных
data = pd.DataFrame({
    'age': [25, 30, 35, 40, 45],
    'salary': [50000, 60000, np.nan, 70000, 80000],
    'city': ['New York', 'Paris', 'Berlin', 'London', 'Paris']
})
```

```python
# Обработка пропусков в целевой переменной (заполняем средним)
data.loc[:, 'salary'] = data['salary'].fillna(data['salary'].mean())
#Используем data.loc[:, 'salary'] для безопасного изменения столбца salary
```

```python
# Целевая переменная (зарплата)
target = data['salary']
features = data.drop(columns=['salary'])
```

```python
# Разделение данных на тренировочные и тестовые
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
# Определяем числовые и категориальные признаки
numerical_features = ['age']
categorical_features = ['city']
```

```python
# Шаги для числовых признаков
numerical_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),  # Заполнение пропусков средним значением
    ('scaler', StandardScaler())                  # Масштабирование данных
])
```

```python
# Шаги для категориальных признаков
categorical_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),  # Заполнение пропусков наиболее частыми значениями
    ('onehot', OneHotEncoder(handle_unknown='ignore'))     # One-hot кодирование категорий
])
```

```python
# Объединяем числовой и категориальный пайплайны в ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_pipeline, numerical_features),
        ('cat', categorical_pipeline, categorical_features)
    ]
)
```

```python
# Создаем общий пайплайн, включающий предобработку и модель машинного обучения
model_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),               # Шаг предобработки данных
    ('model', LinearRegression())                 # Модель линейной регрессии
])
```

```python
# Обучение модели на тренировочных данных
model_pipeline.fit(X_train, y_train)
```

```python
# Предсказание на тестовых данных
y_pred = model_pipeline.predict(X_test)
```

```python
# Оценка модели (например, среднеквадратичная ошибка)
mse = mean_squared_error(y_test, y_pred)
print(f'Среднеквадратичная ошибка (MSE): {mse}')
```

```python
# Выводим предсказанные значения и фактические
print("Предсказанные значения:", y_pred)
print("Фактические значения:", y_test.values)
```
